```python
%%bigquery

#this query uses outer join on the 2015 and 2016 tables based on facilities that don't provide substance abuse treatment
select s1.lst, s1.caseid
from samhda_staging.samhda_2015 as s1
full outer join samhda_staging.samhda_2016 as s2
on s1.treatmt = s2.treatmt where s1.treatmt = 0 and s2.treatmt = 0
limit 12
```

```python
%%bigquery

#this query uses outer join on the 2017 and 2018 tables based on if the facility focuses on mental health treatment. we ordered it by 
#facility type
select s1.facilitytype
from samhda_staging.samhda_2017 as s1
full outer join samhda_staging.samhda_2018 as s2
on s1.focus = s2.focus where s1.focus = 1 and s2.focus = 1 order by s1.facilitytype
limit 12

```

```python
%%bigquery

#this query inner joins the 2017 and 2018 tables based on facilities that don't offer integrated primary care services, 
#and we listed it by publicagency. 

select s1.publicagency
from samhda_staging.samhda_2017 as s1
inner join samhda_staging.samhda_2018 as s2
on s1.primarycare = s2.primarycare where s1.primarycare = 0 and s2.primarycare = 0
limit 12
```

```python
%%bigquery

#this query uses inner join on the 2012 table and 2015 table based on facilities that don't provide mental health treatment in a 
#24-hour hospital inpatient setting. we listed it by the childad variable.

select s1.childad
from samhda_staging.samhda_2012 as s1
inner join samhda_staging.samhda_2015 as s2
on s1.settingip = s2.settingip where s1.settingip = 0 and s2.settingip = 0
limit 12
```

```python
%%bigquery

#this query uses self join on the 2014 table based on facilities that don't tailor their programs for people with 
#Alzheimer's or dementia, and we listed it by the variable "seniors"

select s1.seniors, s2.alzhdementia
from samhda_staging.samhda_2014 as s1
join samhda_staging.samhda_2014 as s2
on s1.alzhdementia = s2.alzhdementia where s1.alzhdementia = 0 and s2.alzhdementia = 0
limit 12
```

```python
%%bigquery

#this query uses self join on the 2015 table based on facilities that do provide substance abuse treatment, and we 
#listed it based off of states and treatment status. 

select s1.lst,s1.treatmt
from samhda_staging.samhda_2015 as s1
join samhda_staging.samhda_2015 as s2
on s1.treatmt = s2.treatmt where s1.treatmt = 1 and s2.treatmt = 1
limit 12
```
