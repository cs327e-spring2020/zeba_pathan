```python
%%bigquery
#these facilities offer substance abuse treatment, in order of states 
select * from samhda_staging.samhda_2012 where treatmt = 1 order by lst
```

```python
%%bigquery
#these facilities offer family psychoeducation, ordered by caseid 
select * from samhda_staging.samhda_2014 where fampsyched = -1 order by caseid
```

```python
%%bigquery
#department of veterans affairs operate these facilities, in order of state and caseid
select * from samhda_staging.samhda_2015 where publicagency = 6 order by caseid,lst
```

```python
%%bigquery
#these mental health treatment centers accept cash or self-payment, ordered by publicagency  
select * from samhda_staging.samhda_2016 where revchk1 = 1 order by publicagency
```

```python
%%bigquery
#these facilities accept patients 65 years or older for treatment, ordered by state
select * from samhda_staging.samhda_2017 where seniors = 1 order by lst
```

```python
%%bigquery
#these facilities do not offer suicide prevention services, and they are in order of caseid
select * from samhda_staging.samhda_2018 where mhsuicide = 0 order by caseid
```
