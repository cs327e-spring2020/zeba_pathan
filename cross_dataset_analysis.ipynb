```python
%%bigquery

#this query joins both facility information tables from our samhda dataset and our ssats dataset
#we did an inner join on the variable "ownershp" which represents what kind of ownership that facility has, and we joined on when 
#when ownership is equal to 1 which means that facility is owned by a private for profit organization
#we used sum to find how many facilities are in each state that are owned by a private for profit organization
#we grouped by lst(state)

select s1.lst, sum(s1.ownershp) as ownershp,
from samhda_modeled.facility_information_Beam_DF as s1
inner join ssats_modeled.ssats_facility_information_Beam_DF as s2
on s1.ownershp = s2.ownershp where s1.ownershp = 1 and s2.ownershp = 1
group by lst
```

```python
%%bigquery

#this query joins both facility information tables from our samhda dataset and our ssats dataset
#we did an inner join on the variable "ownershp" which represents what kind of ownership that facility has, and we joined on when 
#when ownership is equal to 2 which means that facility is owned by a private non profit organization
#we used count to find how many facilities are in each state that are owned by a private non profit organization
#we grouped by lst(state)

select s1.lst, count(*) as ownershp,
from samhda_modeled.facility_information_Beam_DF as s1
inner join ssats_modeled.ssats_facility_information_Beam_DF as s2
on s1.ownershp = s2.ownershp where s1.ownershp = 2 and s2.ownershp = 2
group by lst

```

```python
%%bigquery

#this query joins both facility treatment tables from our samhda dataset and our ssats dataset
#we did an inner join on the variable "treatmt" when treatmt is equal to "Yes" which is whether the facility offers substance abuse treatment
#for the ssats dataset, or whether the facility offers mental health treatment for our samhda dataset
#we used count to find how many facilities offer both of those services
#we grouped by group therapy to see whether they had group therapy or not 

select s1.treatgrpthrpy, count(*) as count,
from samhda_modeled.facility_treatment_Beam_DF as s1
inner join ssats_modeled.ssats_facility_treatment_Beam_DF as s2
on s1.treatmt = s2.treatmt where s1.treatmt = "Yes" and s2.treatmt = "Yes"
group by s1.treatgrpthrpy



```

```python
%%bigquery

#this query joins both facility information tables from our samhda dataset and our ssats dataset
#we did an inner join on the variable "ownershp" which represents what kind of ownership that facility has, and we joined on when 
#when ownership is equal to 1 which means that facility is owned by a private for profit organization
#we used sum to find how many facilities are in each state that are owned by a private for profit organization
#we grouped by lst(state)

create or replace view ssats_modeled.v_ssats_private_forprofit_facilities as
select s1.lst, sum(s1.ownershp) as ownershp,
from `first-planet-266901.samhda_modeled.facility_information_Beam_DF` as s1
inner join `first-planet-266901.ssats_modeled.ssats_facility_information_Beam_DF` as s2
on s1.ownershp = s2.ownershp where s1.ownershp = 1 and s2.ownershp = 1
group by lst
```
