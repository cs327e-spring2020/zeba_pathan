```python
%%bigquery
#we chose to sum the settingop variable (which represents facilities that provide mental health treatment in an
#outpatient setting) and we did a self join with the facility_information_Beam_DF table. We grouped and ordered it by state

select s1.lst, sum(s1.settingop) as settingop,
from samhda_modeled.facility_information_Beam_DF s1 join samhda_modeled.facility_information_Beam_DF s2
on s1.ownershp = s2.ownershp
group by s1.lst
order by s1.lst
```

```python
%%bigquery
#we chose to sum the treattraumathrpy variable (which represents facilities that offer trauma therapy)
#and we did a self join with the facility_information_Beam_DF table. We grouped and ordered it by caseid

select s1.caseid, sum(s1.treattraumathrpy) as treattraumathrpy,
from samhda_modeled.facility_treatment_Beam s1 join samhda_modeled.facility_treatment_Beam s2
on s1.treatmt = s2.treatmt
group by s1.caseid
order by s1.caseid
```

```python
%%bigquery
#we chose to sum the primarycare variable (which represents facilities that offer integrated primary care services)
#and we did a self join with the facility_services_Beam table. We grouped and ordered it by caseid

select s1.caseid, sum(s1.primarycare) as primarycare,
from samhda_modeled.facility_services_Beam s1 join samhda_modeled.facility_services_Beam s2
on s1.mhchronic = s2.mhchronic
group by s1.caseid
order by s1.caseid
```

```python
%%bigquery
#we chose to sum the childad variable (which represents facilities that accept children aged 12 years or younger for treatment)
#from the facility_information_Beam_DF table. We grouped and ordered it by state

select lst, sum(childad) as childad,
from samhda_modeled.facility_information_Beam_DF
group by lst
order by lst
```

```python
%%bigquery
#we chose to sum the settingop variable (which represents facilities that provide mental health treatment in an outpatient
#setting) from the facility_information_Beam_DF table. We grouped by state

select lst, sum(settingop) as settingop,
from samhda_modeled.facility_information_Beam_DF
group by lst
having lst = "TX" or lst = "FL" or lst = "CA" or lst = "NY"
```

```python
%%bigquery
#we chose to sum the childad variable (which represents facilities that accept children aged 12 or younger) from the 
#facility_information_Beam_DF table. we grouped by facility type
select facilitytype, sum(childad) as childad,
from samhda_modeled.facility_information_Beam_DF
group by facilitytype
having facilitytype = 7 or facilitytype = 8 or facilitytype = 1
```

## subqueries


```python
%%bigquery

#we chose to sum the childad variable (which represents facilities that accept children aged 12 or younger) from 
#the facility information table where the caseids used were the ones that had Texas as their state (as shown by the subquery)

select sum(childad) as childad
from samhda_modeled.facility_information_Beam_DF
where caseid IN (select caseid from samhda_modeled.facility_information_Beam_DF where lst = "TX")
```

```python
%%bigquery
#we did a self join on the facility services table on the variable mhchronic (chronic disease/illness management), and we selected entries with caseids that had 
#the primarycare variable = 1, which meant that those facilities offer integrated primary care services.
#we grouped by and ordered by caseid.


select s1.caseid
from samhda_modeled.facility_services_Beam_DF s1 join samhda_modeled.facility_services_Beam_DF s2
on s1.mhchronic = s2.mhchronic
where s1.caseid IN (select caseid from samhda_modeled.facility_services_Beam_DF where primarycare = 1)
group by s1.caseid
order by s1.caseid
```

```python
%%bigquery
#we did a self join on the facility information table on the variable ownershp (which represents whether facilities are 
#owned by a private non profit organization) on those entries where childad = 1 (meaning those facilities accepted children
#aged 12 or younger). we grouped by and ordered by caseid.

select s1.caseid
from samhda_modeled.facility_information_Beam_DF s1 join samhda_modeled.facility_information_Beam_DF s2
on s1.ownershp = s2.ownershp
where s1.caseid IN (select caseid from samhda_modeled.facility_information_Beam_DF where childad = 1)
group by s1.caseid
order by s1.caseid
```

```python
%%bigquery

#in this query we summed the settingop variable (which represents facilities that provide mental health treatment in an outpatient
#setting) from the facility information table, and the caseids chosen were the ones where the facilities
#were in Texas. we grouped by facilitytype. 

select facilitytype, sum(settingop) as settingop,
from samhda_modeled.facility_information_Beam_DF
where caseid IN (select caseid from samhda_modeled.facility_information_Beam_DF where lst = "TX")
group by facilitytype
```

```python
%%bigquery
#this creates the database view for our last query that we included. 

create or replace view samhda_modeled.v_outpatient_facilities as
select facilitytype, sum(settingop) as settingop,
from `first-planet-266901.samhda_modeled.facility_information_Beam_DF`
where caseid IN (select caseid from `first-planet-266901.samhda_modeled.facility_information_Beam_DF` where lst = "TX")
group by facilitytype
```
