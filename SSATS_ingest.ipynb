### create unique id for each table to serve as pk and fk before unioning the tables 

```python
%%bigquery

CREATE TABLE ssats_staging.ssats_2018_u AS
SELECT generate_uuid() as CASEID, *
FROM ssats_staging. ssats_2018
WHERE CASEI_3 IS NOT NULL
```
