```python
%%bigquery
#this query joins the coronabycountry and coronabystate datasets on the dates
#this allows us to see on each day how many people died due to covid-19 summed up 
#by overall country, and in the U.S., how many people died total throughout the country.

select sum(s1.deaths) as deathsbycountry,sum(s2.deaths) as deathsbystate,
from extracreddataset1.coronabycountry as s1
inner join extracreddata2.coronabystate as s2
on cast(s1.date as String) = cast(s2.date as String) where s1.date = "2020-03-15" and s2.date = "2020-03-15"
group by countries_and_territories
```
